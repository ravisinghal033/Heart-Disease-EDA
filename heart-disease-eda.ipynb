```python

import os
import numpy as np
import pandas as pd 
import seaborn as sns

for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df = pd.read_csv('/kaggle/input/heart-failure-prediction/heart.csv')
```

## DISTRIBUTION PLOT for Numeric Columns 

```python
sns.displot(x = df['Age'], kde = False)
```

```python
sns.distplot(x = df['Age'], kde = True, hist = True , rug = True)
```

```python
sns.distplot(x = df['Age'],bins = 20)
```

```python
sns.distplot(x = df['RestingBP'], kde = True, hist = True , rug = True)
```

```python
sns.distplot(x = df['MaxHR'], kde = True, hist = True , color = 'blue')
```

## Categorical Columns (Pie Charts)

```python
df.groupby('Sex').size().plot(kind = 'pie', autopct = '%0.2f')
```

```python
df.groupby('ChestPainType').size().plot(kind = 'pie',autopct = '%0.2f')
```

```python
df.groupby('HeartDisease').size().plot(kind = 'pie',autopct = '%0.2f')
```

## ViolinPlot

```python
sns.violinplot( x = df['Age'])
```

```python
sns.violinplot(y = df['Sex'], x = df['HeartDisease'] , data = df)
```

```python
sns.violinplot(y = df['Cholesterol'], x = df['HeartDisease'] , data = df)
```

```python
sns.violinplot(y = df['Age'], x = df['RestingECG'] , data = df)
```
